```python
import pandas as pd
import numpy as np
```

```python
db = pd.read_csv('../input/dragonball-dataset/Dragon_Ball_Data_Set.csv', encoding='utf-8')
print(db.head(10))
```

```python
print(db.columns)
```

```python
print(db.index)
```

```python
db.rename(index=str, columns={
    'Power_Level': 'Power',
    'Saga_or_Movie': 'Saga',
    'Dragon_Ball_Series': 'Serie'},
    inplace=True)
```

```python
print(db.head(10))
```

```python
db_null = db.isnull().sum().sort_values(ascending=False)
print(db_null)
```

```python
db = db.dropna()
print(db.head(10))
```

```python
db.drop_duplicates()
```

```python
db['Power'] = db['Power'].str.replace(',','')
```

```python
db['Power'] = db['Power'].str.replace(',','.')
print(db.head(10))
```

```python
db['Power'] = pd.to_numeric(db['Power'], errors='ignore')
print(db.tail(10))
```

```python
db.to_csv('DragonBall_Datafinal.csv', index = False)
```
